```python
# Text Adventure Game ##
## Data Science: Python ##
## Hult International Business School ##
## The Maze Runner 2014 Sci-fi/Action Film##
"""
Created on Sunday, October 24, 2021

@author: Krystel Lualhati

"""

"""
    DocString:
    
    A)Introduction: 
    This game is a dystopian novel based on the famous book series 
    "The Maze Runner" by James Dashner, then later adapted into a film 
    in 2014. The story begins with Thomas waking up finding himself in 
    the center of a giant labyrinth with no memory of his past. Then later, 
    his character developed him to be one of the fastest runners 
    to find their way out of the maze. 
    
    In this game, the main character will take you a glimpse of the 
    thrilling adventure of Thomas in the maze in three different stages:
    
    Stage 1: The Maze - Escape the griever and hide Alby
    Stage 2: The Griever - Explore the unknown of the maze. 
    Stage 3: The Sections - Find Alby and get back to the Glade.
    
    Movie Reference: 32:00 - 41:00
    
    B) Known Bugs and/or Errors: 2
    1 - TypeError: 'str' object is not callable - Try Again
    
"""

#First scene in the movie
def day_one():
    print(f"""
    -DAY ONE-
    {'-' * 40}

    You are in a confined space, pitch black. Slowly, shapes and boundaries 
    emerge as a dim room, the size of a large freight elevator, 
    gradually comes into a focus.\n
    A teenage boy is awakening...,and that's you.
    """)
    input(prompt = "\n<Press enter to continue>\n")
    identity() #this will proceed to the next phase

#Identify yourself as the teenage boy.
#Providing option to create your own character

def identity():
    print("""
    The only thing you could remember about your life is your name.\n 
    Would you like to play as Thomas, or your own character?
    """)
#prompt the choices
#else invalid command to continue
#inserting nested conditional statement
    choice = input(prompt = """1) play as Thomas or 2) create my own character
                   \n""")
    #prompting choices while triggering lower and capitalize to meet the requirements
    #if true, it will proceed to get started
    if "1" in choice or "play" in str.lower(choice) or "Thomas" in str.capitalize(choice):
        print(f"""\n 
        Nice to meet ya, Thomas! Welcome to the Glade!
        \U0001F941\U0001F941\U0001F941\U0001F941\U0001F941
        """)
        input(prompt ="\n<Press enter to continue>\n")
        get_started() 
    #if true, it will proceed to character building
    elif "2" in choice or "create" in str.lower(choice) or "character" in str.lower(choice):
        print("\nOkay, shank! Build your own character\n")
        character_build()
    else:
        print("Invalid command. Please try again.")
        input(prompt ="\n<Press enter to try again>\n")
        identity() #This goes back to first function
        
#Asking the user to input the name
#Applying capitalize to trigger upper case for teenage boy
def character_build():
    teenage_boy = input(prompt = 
                    "\nWhat name do you remember, shank?\n").capitalize()
    if teenage_boy:
        print(f"""
        Nice to meet ya, {teenage_boy}! Welcome to the Glade.\n
        \U0001F941\U0001F941\U0001F941\U0001F941\U0001F941
        
        """)
        input(prompt ="\n<Press enter to continue>\n")
        get_started()
    #No input will create a new command
    else: 
        print(f"""
        You forgot your name also?!
        Don't worry! We'll call you Thomas!
        """)
        input(prompt ="\n<Press enter to continue>\n")
        print(f"""\n 
        Nice to meet ya, Thomas! Welcome to the Glade!
        \U0001F941\U0001F941\U0001F941\U0001F941\U0001F941
        """)
        input(prompt ="\n<Press enter to continue>\n")
        get_started()
    
#The start of the game
#If user fails, the user will lead back to this stage to skip character creation
def get_started():
#Creating list to save answers for prompt
    start_lst = []
    
    #Asking if the User is ready to start
    #Starting a for loop
    get_started = input(prompt = "Let's Get Running? [Y/N]\n")
    
    #Inserting answer in list
    start_lst.append(get_started)
    
    #evaluating the answer while UPPER, to see if it matches the requirement
    #applying try and except for yes and no
    for element in start_lst:
        try: 
            ['Yes', 'Yas', 'yess', 'Nooo','nah']
            #except nameerrors if input integers or skips enter
        except NameError:
            print('letters only')
        #if true, then it will start story plot, if not, it will ask again  
        if 'Y' in str.upper(element) or 'y' in str.lower(element):
            print(f"\nWarm up your legs. Let's Go!\n\n")
            the_maze()
        #if no, the user will be asked to run the code again    
        elif 'N' in str.upper(element) or 'n' in str.lower(element):
            print(f"""
            \nYou may have outrun this game, but you cannot outrun 
            starting the code again!\n\n
                  """)
        else:
            print(f"\nYour answer is invalid, please try again.\n\n") 
            input(prompt ="\n<Try Again>\n")
            get_started() #calling the previous function to repeat the process
            
#The start of the game
#The story plot
def the_maze():
    print(f"""
    -THE MAZE-
    {'-' * 40}
    You notice that each of the four cardinal direction the wall split open
    to reveal a spiraling labyrinth of twisting and turning passages that 
    unravel around the Glade forever into the distance... the Maze.\n
    {'-' * 40}
    """)
    input(prompt ="<Press enter to continue>\n")
    print(f"""
    The bell rangs indicating the Walls are about to begin closing. 
    You looked around and spots all Runners, except Minho and Alby.
    If they're not back now, they're goners. From the inside the Maze, 
    you spotted Minho and Alby, but Alby is wounded.\n
    """)
    input(prompt ="<Press enter to continue>\n")
    run_thomas()

#Running to the Maze
#Prompting the first decision 
#importing random to be applied in a nested conditional statement
import random
def run_thomas():
    print("""
    A whisper in your head..."Run..." But Gally stopped you. 
    If you run towards the Maze now, YOU ARE A GONER. 
    """)
    #First Decision 
    #Applying Conditional Statement for Level 1
    choice = input(prompt = f"""
    What are you going to?\n
    1.) ignore Gally to help Minho.
    2.) stay because it's a Glade rule that you cannot enter the Maze.\n
    """)
    #If 1, then it will proceed to the next decision making.
    #Using .lower and .capitalize to meet the requirements
    if "1" in choice or "help" in str.lower(choice) or "Gally" in str.capitalize(choice):
        print("""
    You raced into the Maze, just as the doors SLAM shut behind you.\n
              """)
        input(prompt ="<Press enter to continue>\n")
        hide_alby()     
    #If not, it will proceed to fail
    elif "2" in choice or "stay" in str.lower(choice) or "Glade" in str.capitalize(choice): 
        fail()    #But you can try again, if input is empty.
    else:
        print(f"""\nCan't decide? Don't be scared!\n""")
        input(prompt ="<Try Again!>\n")
        get_started()
        
#Inside the Maze 
#The end goal is to hide alby
#Decision 2 choices
def hide_alby():
    print(f"""
    -INSIDE THE MAZE-
    {'-' * 40}
    From somewhere out in the Maze, you heard a loud shriek...the Griever.
    Alby is close to dying, and you need to hide.
    """)
    choice = input(prompt ="""
    What are you going to do next?
    1.) Leave Alby and hide with Minho. 
    2.) Keep dragging Alby\n
    """)
    #the fail function will return back to get_started
    #Using .lower and .capitalize to meet the requirements
    if "1" in choice or "Leave" in str.capitalize(choice) or "hide" in str.lower(choice):
        print("""You let your friend, Alby die.\n
              """)
        input("<Try Again>")
        hide_alby() #this goes back to hide_alby function
     
    elif "2" in choice or "Keep" in str.capitalize(choice) or "Alby" in str.capitalize(choice):
        print("""
    You found a vine wall - While you are tying Alby to hide 
    from the Griever.You can hear the Griever is getting closer 
    and Minho ran away. 
        """)
    
        #insert if and while function to make it more interactive
        #this will a prompt a question to keep playing or not
        decision = input(prompt = """
    It's also time for you to run away.
    1.) no, keep pulling the vine.
    2.) yeah, time to follow Minho.
        \n""")
        #applying while to finish the pulling
        #Using .lower and .capitalize to meet the requirements
        if "1" in decision or "no" in str.lower(decision) or "vine" in str.lower(decision):
            print("Another five feet up, press enter to pulling.\n")
            presses = 5
            # keep pulling
            while presses > 0:
                print(presses)
                presses = presses -1
                input(prompt ='<Keep pulling>\n')
            print('That was closed! You saved, Alby! Great job!')
            input("<Press enter to continue>")
            the_griever()
        #if true, it prompt to try again the game instead of fail.        
        elif "2" in decision or "follow" in str.lower(decision) or "Minho" in str.capitalize(decision):
            print("""
    You may have saved yourself from the Griever, but you left your friend...
            """)
            input(prompt ="<Try Again>")
            hide_alby()
        #else applies if input is empty
        else:
            print(f"""\nCan't decide? Don't be scared!\n""")
            input(prompt ="<Try Again!>\n")
            hide_alby()
            
#The start of round 2
#importing random direction
def the_griever():
    import random
    #declaring objects
    #converting objects into integers
    left = int(1)
    right = int(2)
    random_direction = random.randint(1,2) #left, right, and straight
    chances = 1
    #looping until the player ran out of chances
    print(f"""
    -THE GRIEVER-
    {'-' * 40}
    After you saved Alby, you kept running. 
    You realized that the maze is a complete unknown to you.
    You're at the center of a T-intersection.
    """)
    #The user only get few chances to guess which way to turn.
    #applying while loop to guess the direction
    while chances > 0:
        direction = input(prompt = """
    Which way would you turn? 
    1. left
    2. right
        """)
        #converting direction for numerical comparison
        direction = int(direction)
        if direction == random_direction:
            print("""
    Fifty yards down, you bumped into Minho. Both of you started
    running as soon as you heard shriek. It's a game of chicken 
    as the Griever sprinting directly towards your direction.
            """)    
            ##Guess another direction
            ## the first guessing question will have few chances
            direction = input(prompt = """
    Where do you go next?
    1. left
    2. right
            """)
            #converting direction for numerical comparison
            direction = int(direction)
            if direction == random_direction:
                    print("""
    As the Griever about to catch up, 
    you leaped through disappearing doorway\n
                    """)
                    print('Now, you have to run. Press enter to run\n')
                    run = 5
                    #Adding while loop to proceed to run
                    while run > 0:
                        print(run)
                        run = run -1
                        input(prompt ='<Keep running>\n')
                    print("""
    As you run through the closing wall, 
    the Griever's head is squeezed,and the wall comes 
    off its tracks, coming to a stop.\n
                    """)
                    input(prompt ="<Press enter to continue>")
                    print(f"""
    That was a nice move back there! 
    You killed a griever! Now, let's find alby.
    \U0001F4AF\U0001F4AF\U0001F4AF\U0001F4AF\U0001F4AF
                    """)
                    input(prompt ="<Press enter to continue>")
                    the_sections()
                    break
                    chances += 1
                    #nested condition elif if fail to guess the number
            elif direction != random_direction:
                print("You're in a dead end.")
                input(prompt ="<Try Again>")
                the_griever()
                       
        #Wrong Guess would bring to a dead end
        #Elif to bring back to start the game again
        elif direction != random_direction:
            print("You're in a dead end. The griever caught you!")
            input(prompt ="<Try Again>")
            the_griever() #Calling the function for round 2
        else:
            print("Think as quick as the griever! Make up your mind fast!")
            input(prompt ="<Invalid Command. Try Again!>\n")
            break
            the_griever()  #Calling the function for round 2
         
#Round 3

def the_sections():
    import random
    #declaring objects
    solution_1 = 5
    solution_2 = 7
    solution_3 = 8
    print(f"""
    -MASSIVE CIRCLES-
    {'-' * 40}
    The map is a massive circle with eight sections heading off in
    in different directions. While running, you need to recall which 
    section you left Alby through solving this final question.
    """)
    #creating a list for random questions
    question_list = ['While jogging through the snaking passage...',
                    'You spotted a small archway ahead...?',
                     'You came across a Cavern...']
    questions = 3
    #Using while loops to prompt the list.
    #Using random choice to prompt random questions
    while questions > 0:
        question = random.choice(question_list)
        input(prompt ="<Press enter to start running>\n")
        print(question)
        print("")
        #prompting question 1 from the list
        if question == question_list[0]:
            print('What section is 3+2?')
            solution_1 = (input(prompt = '\n'))
            
            if '5' in solution_1 or 'five' in solution_1:
                print("MINHO! It's section 5!\n")
                questions -=1
                
            elif (solution_1) != input:
                print("""
                You're too late. The stung from the Griever got Alby.
                DAMN! You are bad at math and recalling numbers.
                """)
                fail()
                
         #prompting question 2 from the list      
        elif question == question_list[1]:
            print('What section is 28/4?')
            solution_2 = (input(prompt ='\n'))
            
            if '7' in solution_2 or 'seven' in solution_2:
                print("MINHO! It's section 7!\n")
                questions -=1
                
            elif (solution_2) != input:
                print("""
                You're too late. The stung from the Griever got Alby.
                DAMN! You are bad at math and recalling numbers.
                """)
                fail()
         #prompting question 3 from the list
        elif question == question_list[2]:
            print('Which section is the square root of 64?')
         
            solution_3 = (input(prompt ='\n'))
            #converting direction for numerical comparison
            
            if '8' in solution_3 or 'eight' in solution_3:
                print("MINHO! It's section 8\n")
                questions -=1
                
            elif solution_3 != input:
                print("""
                You're too late. The stung from the Griever got Alby.
                DAMN! You are bad at math and recalling numbers.
                """)
                fail() #Calling fail function
        
        else: #inputing else to if all else fail
            print("DAMN! You are bad at math and recalling numbers.")
            input(prompt ="<Try Again!>\n")
            the_sections()
            
        #Seeing the end of the game      
        print("""
Alby's foot is barely visible through the vegetation, and
you can hear the wall's bolts retract to open.
              """)
        input(prompt ="<Press enter to go back to the Glade>")
        the_glade() #goes to the exit function to end the game.
        break
              
#indicates the end of the game
#defining exit function
def the_glade():
    print("""
You have made it back to the Glade! Congratulations!
\U0001F941\U0001F941\U0001F941\U0001F941\U0001F941
        """)
    input(prompt = "<Press enter to play again>")
    day_one() #calling the first function to start the game
                
#inserting a fail function
#return to get_started 
def fail():
    print('\n\nGAME OVER! You have failed to escape the Maze.\n\n')
    input(prompt ="<Press enter to start again>")
    get_started()  #adding exit function to end the game or to try again
        
day_one()
```
